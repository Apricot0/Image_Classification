```python
# import packages here
import os
import sys
print (sys.executable)
import cv2
import numpy as np

import matplotlib.pyplot as plt
import glob
import random
import time

import torch
import torchvision
import torchvision.transforms as transforms

from torch.autograd import Variable
from torch import Tensor
from typing import List, Optional
import torch.nn as nn
import torch.nn.functional as F
```

```python
def set_random_seed(seed):
     torch.manual_seed(seed)
     torch.cuda.manual_seed_all(seed)
     np.random.seed(seed)
     random.seed(seed)
     torch.backends.cudnn.deterministic = True
set_random_seed(0)

class_names = [name[13:] for name in glob.glob('./data/train/*')] # name attributes
class_names = dict(zip(range(len(class_names)), class_names))
print("class_names: %s " % class_names)

def img_rot(img):
  if len(img.shape) == 2 :
        degree = random.uniform(-30, 30)
        rotation_matrix = cv2.getRotationMatrix2D((img.shape[1]/2, img.shape[0]/2), degree, 1)
        rotated_img = cv2.warpAffine(img, rotation_matrix, (img.shape[1], img.shape[0]))
        return rotated_img
  elif len(img.shape) == 3:
        # Rotate each channel of the RGB image by 20 degrees
        rotated_channels = []
        for channel in img:
            rotation_matrix = cv2.getRotationMatrix2D((channel.shape[1]/2, channel.shape[0]/2), 20, 1)
            rotated_channel = cv2.warpAffine(channel, rotation_matrix, (channel.shape[1], channel.shape[0]))
            rotated_channels.append(rotated_channel)
        rotated_img = np.stack(rotated_channels, axis=0)
        return rotated_img

# split data into train and test
def load_dataset(path, img_size, num_per_class=-1, batch_size=16, shuffle=False,
           augment=False, zoom=False, rotate=False, crop=False, is_color=False, zero_centered=False):
    data = []
    labels = []
    channel_num = 3 if is_color else 1

    # read images and resizing
    for id, class_name in class_names.items():
        print("Loading images from class: %s" % id)
        img_path_class = glob.glob(path + class_name + '/*.jpg')
        if num_per_class > 0:
            img_path_class = img_path_class[:num_per_class]
        labels.extend([id]*len(img_path_class))
        for filename in img_path_class:
            if is_color:
                img = cv2.imread(filename)
            else:
                img = cv2.imread(filename, 0)

            # resize the image
            img = cv2.resize(img, img_size, cv2.INTER_LINEAR)

            if is_color:
                img = np.transpose(img, [2, 0, 1])

            # norm pixel values to [-1, 1]
            data.append(img.astype(np.float64)/255*2-1)
    if augment:
        data.extend([img_rot(img) for img in data])
        labels.extend(labels)
        # data.extend([img_rot(img_rot(img)) for img in data])
        # labels.extend(labels)
        data.extend([np.fliplr(img) for img in data])
        labels.extend(labels)
        # print("data after augmentation: ", data)

    # Data Normalization
    if zero_centered:
        # for img in data:
        #     img -= np.mean(img)
        #     # img /= np.max(img) - np.min(img)
        data -= np.mean(data, 0)

    # randomly permute (this step is important for training)
    if shuffle:
        bundle = list(zip(data, labels))
        random.shuffle(bundle)
        data, labels = zip(*bundle)

    # divide data into minibatches of TorchTensors
    if batch_size > 1:
        batch_data = []
        batch_labels = []

        for i in range(int(len(data) / batch_size)):
            minibatch_d = data[i*batch_size: (i+1)*batch_size]
            minibatch_d = np.reshape(minibatch_d, (batch_size, channel_num, img_size[0], img_size[1]))
            batch_data.append(torch.from_numpy(minibatch_d))

            minibatch_l = labels[i*batch_size: (i+1)*batch_size]
            batch_labels.append(torch.LongTensor(minibatch_l))
        data, labels = batch_data, batch_labels
    return zip(batch_data, batch_labels)
```

```python
# reload data with a larger size
img_size = (64, 64)
batch_size = 16 # training sample number per batch

# load training dataset
trainloader_large = list(load_dataset('./data/train/', img_size, batch_size=batch_size, shuffle=True,
                                      augment=True, is_color=False, zero_centered=True))
train_num = len(trainloader_large)
print("Finish loading %d minibatches (batch_size=%d) of training samples." % (train_num, batch_size))

# load testing dataset
testloader_large = list(load_dataset('./data/test/', img_size, num_per_class=50, batch_size=batch_size, is_color=False, zero_centered=True))
test_num = len(testloader_large)
print("Finish loading %d minibatches (batch_size=%d) of testing samples." % (test_num, batch_size))
```

```python
#--------------------------------------------------
#       Model Training Function
#--------------------------------------------------
import torch.optim as optim
import time

def trainModel(net, trainloader, train_option, testloader=None):
  loss_func = nn.CrossEntropyLoss()
  lr = train_option['lr']
  epoch = train_option['epoch']
  device = train_option['device'] if 'device' in train_option.keys() else 'cpu'
  log_iter = train_option['log_iter'] if 'log_iter' in train_option.keys() else 20
  eval_epoch = 1

  if 'optimizer' in train_option.keys():
    optimizer = train_option['optimizer']
  else:
    optimizer = optim.Adam(net.parameters(), lr=lr)

  start_time = time.time()
  if device == 'gpu':
    net = net.cuda()

  iters = 0
  running_loss = 0.0
  for ep in range(epoch):
    net.train()
    for iter, (x, y) in enumerate(trainloader):
      iters += 1
      batch_x = Variable(x).float()
      batch_y = Variable(y).long()
      # print(batch_x.shape)
      # print(batch_y.shape)
      if device == 'gpu':
        batch_x = batch_x.cuda()
        batch_y = batch_y.cuda()

      outputs = net(batch_x)
      loss = loss_func(outputs, batch_y)
      optimizer.zero_grad()
      loss.backward()
      optimizer.step()
      running_loss += loss.item()

      time_lapse = time.strftime('%H:%M:%S', time.gmtime(time.time() - start_time))
      if iter % log_iter == 0:
        print('Epoch:{:2d} | Iter:{:5d} | Time: {} | Train Loss: {:.4f} | Average Loss: {:.4f} '.format(ep+1, iter, time_lapse, loss.item(), running_loss/iters))

    if testloader is not None and ep % eval_epoch == 0:
      evalModel(net, testloader)
```

```python
#--------------------------------------------------
#       Model Evaluating Function
#--------------------------------------------------
import time

def evalModel(net, testloader):
  acc = 0.0
  count = 0
  start_time = time.time()
  device = 'gpu' if next(net.parameters()).is_cuda else 'cpu'
  net.eval()

  for iter, (x, y) in enumerate(testloader):
        count += x.shape[0]
        batch_x = Variable(x).float()
        batch_y = Variable(y).long()
        if device == 'gpu':
          batch_x = batch_x.cuda()
          batch_y = batch_y.cuda()
        outputs = net(batch_x)
        acc += torch.sum(outputs.max(1)[1]==batch_y)

  time_lapse = time.strftime('%H:%M:%S', time.gmtime(time.time() - start_time))
  print('Accuracy: {:5f} | Time: {}'.format(acc/count,time_lapse))
```

```python
class PVT(nn.Module):
    def __init__(self,
                 img_size=64,
                 in_channels=1,
                 patch_sizes=[8, 16, 32],
                 embedding_dims=128,
                 num_transformer_layers=2,
                 dropout=0.1,
                 mlp_hidden_dim=128,
                 num_heads=2,
                 num_classes=22):
        super().__init__()
        self.img_size = img_size
        self.in_channels = in_channels

        # Patchify layers for multiple scales
        self.patchify_layers = nn.ModuleList([
            nn.Conv2d(in_channels, embedding_dims, kernel_size=patch_size, stride=patch_size)
            for patch_size in patch_sizes
        ])

        # Determine the number of patches for each scale
        self.num_patches = [(img_size // patch_size) ** 2 for patch_size in patch_sizes]

        # Positional encoding for each scale
        self.positional_encoding = nn.ParameterList([
            nn.Parameter(torch.randn(1, num_patches + 1, embedding_dims))
            for num_patches in self.num_patches
        ])

        # Extra learnable [CLASS] embedding
        self.cls_token = nn.Parameter(torch.randn(1, 1, embedding_dims))

        # Transformer encoder layers for each scale
        self.transformer_encoder_layers = nn.ModuleList([
            nn.ModuleList([
                nn.TransformerEncoderLayer(embedding_dims, num_heads, mlp_hidden_dim, dropout, batch_first=True)
                for _ in range(num_transformer_layers)
            ])
            for _ in range(len(patch_sizes))
        ])

        # MLP head for classification
        self.mlp_head = nn.Sequential(
            nn.Linear(len(patch_sizes) * embedding_dims, mlp_hidden_dim),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(mlp_hidden_dim, num_classes)
        )

    def forward(self, x):
        bs, c, h, w = x.shape
        embeddings = []

        # Patchify and process each scale
        for i, patchify_layer in enumerate(self.patchify_layers):
            x_scale = patchify_layer(x)
            x_scale = x_scale.flatten(2).transpose(1, 2)  # (batch_size, num_patches, embedding_dims)

            # Concatenate cls_tokens and x along the batch dimension
            x_scale = torch.cat((x_scale, self.cls_token.repeat(bs, 1, 1)), dim=1)

            # Add positional encoding
            x_scale = x_scale + self.positional_encoding[i]

            # Transformer encoder layers
            for layer in self.transformer_encoder_layers[i]:
                x_scale = layer(x_scale)

            # Extract class token and append to list of embeddings
            cls_token = x_scale[:, 0]
            embeddings.append(cls_token)

        # Concatenate embeddings from all scales
        x = torch.cat(embeddings, dim=1)

        # Pass through MLP head for classification
        x = self.mlp_head(x)

        return x
```

```python
class ModifiedResNet(nn.Module):
    def __init__(self, original_model):
        super(ModifiedResNet, self).__init__()
        # Load the pre-trained ResNet model
        self.features = nn.Sequential(*list(original_model.children())[:-2])
        # Modify the first convolutional layer to accept single-channel input
        self.features[0] = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)
        # Adjust the weights of the new first convolutional layer
        with torch.no_grad():
            self.features[0].weight[:, :3] = original_model.conv1.weight.mean(dim=1, keepdim=True)
        self.linear = nn.Linear(2048, 1, 3)

    def forward(self, x):
        x = self.features(x)
        x = self.linear(x)
        return x
```

```python
import torchvision.models as models
class PVT(nn.Module):
    def __init__(self,
                 img_size=64,
                 in_channels=1,  # ResNet usually takes 3 channels
                 resnet_layers=4,  # Number of ResNet layers to use as features
                 patch_sizes=[8, 16, 32],
                 embedding_dims=128,
                 num_transformer_layers=2,
                 dropout=0.1,
                 mlp_hidden_dim=128,
                 num_heads=2,
                 num_classes=22):
        super().__init__()
        self.img_size = img_size
        self.in_channels = in_channels

        # ResNet as pre-processing
        self.resnet = ModifiedResNet(models.resnet50(pretrained=True))

        # Determine the number of patches for each scale
        self.num_patches = [(img_size // patch_size) ** 2 for patch_size in patch_sizes]

        # Positional encoding for each scale
        self.positional_encoding = nn.ParameterList([
            nn.Parameter(torch.randn(1, num_patches + 1, embedding_dims))
            for num_patches in self.num_patches
        ])

        # Extra learnable [CLASS] embedding
        self.cls_token = nn.Parameter(torch.randn(1, 1, embedding_dims))

        # Transformer encoder layers for each scale
        self.transformer_encoder_layers = nn.ModuleList([
            nn.ModuleList([
                nn.TransformerEncoderLayer(embedding_dims, num_heads, mlp_hidden_dim, dropout, batch_first=True)
                for _ in range(num_transformer_layers)
            ])
            for _ in range(len(patch_sizes))
        ])

        # MLP head for classification
        self.mlp_head = nn.Sequential(
            nn.Linear(len(patch_sizes) * embedding_dims, mlp_hidden_dim),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(mlp_hidden_dim, num_classes)
        )

    def forward(self, x):
        bs, c, h, w = x.shape
        embeddings = []

        # Pre-process through ResNet
        x = self.resnet(x)
        print(x.shape)
        # Patchify and process each scale
        for i, patch_size in enumerate([8, 16, 32]):  # Assuming patch_sizes are predefined
            num_patches = (self.img_size // patch_size) ** 2
            x_scale = x.clone()  # You can modify this based on your patching strategy
            print(x_scale.shape)
            # Concatenate cls_tokens and x along the batch dimension
            x_scale = torch.cat((x_scale, self.cls_token.repeat(bs, 1, 1)), dim=1)
            # Add positional encoding
            x_scale = x_scale + self.positional_encoding[i]

            # Transformer encoder layers
            for layer in self.transformer_encoder_layers[i]:
                x_scale = layer(x_scale)

            # Extract class token and append to list of embeddings
            cls_token = x_scale[:, 0]
            embeddings.append(cls_token)

        # Concatenate embeddings from all scales
        x = torch.cat(embeddings, dim=1)

        # Pass through MLP head for classification
        x = self.mlp_head(x)

        return x
```
